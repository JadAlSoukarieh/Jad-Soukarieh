```python
from IPython.display import display, clear_output, HTML
def check_winner(b):
    wins = [(0,1,2),(3,4,5),(6,7,8),
            (0,3,6),(1,4,7),(2,5,8),
            (0,4,8),(2,4,6)]
    for x,y,z in wins:
        if b[x]==b[y]==b[z]!=" ":
            return b[x]
    if all(s!=" " for s in b):
        return "Tie"
    return None

def minimax(b, player):
    winner = check_winner(b)
    if winner == "X": return {'score':1}
    if winner == "O": return {'score':-1}
    if winner == "Tie": return {'score':0}
    moves=[]
    for i in range(9):
        if b[i]==" ":
            b[i]=player
            score = minimax(b, "O" if player=="X" else "X")['score']
            moves.append({'index':i,'score':score})
            b[i]=" "
    return max(moves, key=lambda m:m['score']) if player=="X" else min(moves, key=lambda m:m['score'])
```

```python
board = [" "]*9
buttons = []

def update_buttons():
    for i,btn in enumerate(buttons):
        btn.description = board[i]

def end_game():
    w = check_winner(board)
    if w:
        msg = "Tie!" if w=="Tie" else f"{w} wins!"
        for b in buttons: b.disabled=True
        clear_output(wait=True)
        display(grid)
        print("\nResult:", msg)
        return True
    return False

def on_button_click(i):
    if board[i]!=" " or end_game(): return
    board[i] = player_symbol
    update_buttons()
    if end_game(): return
    ai_move = minimax(board, ai_symbol)['index']
    board[ai_move] = ai_symbol
    update_buttons()
    end_game()

for i in range(9):
    btn = widgets.Button(
        description=" ",
        layout=widgets.Layout(
            width='60px', height='60px',
            border='2px solid black', margin='2px'
        ),
    )
    btn.on_click(lambda _, i=i: on_button_click(i))
    buttons.append(btn)

grid = widgets.GridBox(
    buttons,
    layout=widgets.Layout(grid_template_columns="repeat(3, 60px)")
)

if random.choice([True, False]):
    ai_symbol, player_symbol = "X", "O"
    board[minimax(board, "X")['index']] = "X"
else:
    ai_symbol, player_symbol = "O", "X"

update_buttons()

display(HTML("""
<style>
  .widget-button {
    background-color: white !important;
    color: blue !important;
  }
</style>
"""))
print(f"You are '{player_symbol}', AI is '{ai_symbol}'. Click a square to play:")
display(grid)
```
